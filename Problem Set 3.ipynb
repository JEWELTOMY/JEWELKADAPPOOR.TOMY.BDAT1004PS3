# Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("us-marriages-divorces-1867-2014.csv")
```

```python
data
```

```python

columns = data.columns.drop(['Marriages','Divorces','Population','Year'])
fig, ax = plt.subplots()
for col in columns:
    ax.plot(data['Year'], data[col])
ax.set_title('US Marriages & Divorces')
ax.set_xlabel('Year')
ax.set_ylabel('Values  per thousand')
ax.legend(columns)
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt
data2= pd.read_csv('us-marriages-divorces-1867-2014.csv')

marry = data2[(data2.Year == 2000) | (data2.Year == 1900) | (data2.Year == 1950)]
marry.plot(kind= 'bar', x='Year', y=['Marriages_per_1000','Divorces_per_1000'], label=['Marriages','Divorces'])
plt.title("number of marriages and divorces between 1900, 1950, and 2000.")
plt.legend(loc='best')
plt.show()

```

# Question 10
Create a scatter plot showing the relationship between the total revenue earned by 
arcades and the number of Computer Science PhDs awarded in the U.S. between 
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year

```python
import pandas as pd
import matplotlib.pyplot as plt


arcade_vs_csphd = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

fig4,ax4 = plt.subplots()
for i in range (len(arcade_vs_csphd['Total Arcade Revenue (billions)'])):
    ax4.scatter(arcade_vs_csphd['Total Arcade Revenue (billions)'][i],arcade_vs_csphd['Computer Science Doctorates Awarded (US)'][i])
ax4.set_title('Total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S.')
ax4.set_xlabel('Arcade Revenue ')
ax4.set_ylabel('Cs PhDs Awarded')
```
